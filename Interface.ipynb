```python
import pandas as pd
import time

def input_data():
  data = []
  inputting = True
  while inputting:
    inp = input("Enter date and time of failure in the format: yyyy-mm-dd hh:mm:ss (or y to quit): ")
    if inp == "y":
      inputting = False
    else:
      data.append(inp)
  return data

def modify_data(filename='network_metrics.csv'):
  data = input_data()
  indices = []
  datesDf = pd.read_csv(filename)[['Timestamp']]
  datesArray = datesDf.values 
  datesList = datesArray.tolist()
  for date in data:
    givenSecs = time.mktime(time.strptime(date, "%Y-%m-%j %H:%M:%S"))
    indices.append(binary_search(datesList, givenSecs))
  csvfile = pd.read_csv(filename)
  print('changed: ') #test
  for ind in indices:
    csvfile.loc[ind, 'isFail'] = True
    print(f'{ind}, ') #test
  csvfile.to_csv(filename, index=False)
  print("Done!") 

def binary_search(arr, val):
  high = len(arr) - 1
  low = 0
  last = 0
  middle = 0
  while low <= high:
    mid = (high + low) // 2
    # print(arr[mid][0]) #test
    midSecs = time.mktime(time.strptime(arr[mid][0], "%Y-%m-%j %H:%M:%S"))
    if midSecs < val:
      last = low
      low = mid + 1
    elif midSecs > val:
      last = high
      high = mid - 1
    else:
      return mid
    middle = mid
  diffLast = abs(val - time.mktime(time.strptime(arr[last][0], "%Y-%m-%j %H:%M:%S")))
  diffMid = abs(val - time.mktime(time.strptime(arr[middle][0], "%Y-%m-%j %H:%M:%S")))
  if diffLast < diffMid:
    return last
  else:
    return middle

def main():
  modify_data()
  
if __name__ == "__main__":
  main()
```
